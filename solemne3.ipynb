Conjunto de datos sobre la diabetes en una población

Describe los datos de los registros médicos de los pacientes  y si tenían la aparición de la diabetes en un plazo de cinco años. Se trata de un problema de clasificación binaria (aparición de diabetes como 1 o no como 0). Las variables de entrada que describen a cada paciente son numéricas y tienen variables de escalas. A continuación se enumeran los ocho atributos del set de datos:

1. Número de embarazos.
2. Concentración de glucosa en plasma a 2 horas en una prueba oral de tolerancia a la glucosa.
3. Presión arterial diastólica (mm Hg).
4. Grosor del pliegue de la piel del tríceps (mm).
5. Insulina sérica de 2 horas (mu U/ml).
6. Índice de masa corporal.
7. Función del pedigrí de la diabetes.
8. Edad (años).
9. Clase, aparición de diabetes en un plazo de cinco años.

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
import pandas as pd
```

```python
#Importamos datos
data_frame = pd.read_csv("pacientes_diabetes.csv")
```

```python
#Insertamos columna con la cabecera de los datos
data_frame.columns=["Embarazos", "Glucosa", "Presión", "Grosor piel", "Insulina", "Masa corporal", "Pedigrí", "Edad", "Clase"]
#Mostramos los primeros registros
data_frame.head()
```

```python
#Iniciamos generadr aleatorio "seed"
seed = 7
numpy.random.seed(seed)

# Cargamos dataset con numpy
df = numpy.loadtxt("pacientes_diabetes.csv", delimiter=",")
```

```python
#Se reparten los datos en data y targets
X = df[:, 0:8]
Y = df[:, 8]
```

```python
#Utilizamos una secuencia de capas con Sequential
model = Sequential()
#En la primera capa utilizamos 8 variables de entrada, referenta a la cabecera que tenemos y 12 neuronas
model.add(Dense(12, input_dim=8, activation='relu'))
#Asignamos un tootal de 3 capas
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#Compilamos el modelo usando el backend de tensorflow
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Entrenamos el modelo, ejecutamos el entrenamiento 150 veces y se evaluan 10 instancias
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
#Evaluamos el modelo, con la misma entrada y salida del entrenamiento
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
#Nos genera una prediccion
```
